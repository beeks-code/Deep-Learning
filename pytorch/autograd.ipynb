```python
x=torch.tensor(3.0,requires_grad=True)
y=x**2
z=torch.sin(y)
z.backward()
x.grad
```

If we calculate gradient of function multiple time it will get add up each time

```python
x = torch.tensor(2.0, requires_grad=True)

for i in range(3):
    y = x**2
    y.backward()
    print(x.grad)
```

To solve we reset value of grad to zero by in-placed operation

```python
x = torch.tensor(2.7, requires_grad=True)

for i in range(3):
    y = x**2
    y.backward()
    print(x.grad)
    x.grad.zero_() ## store zero at x.grad
```

#### To get gradient of Intermediate Variable

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**2
z = torch.sin(y)

y.retain_grad()  #  tell PyTorch to save grad for y
z.retain_grad()  #  tell PyTorch to save grad for z
"""  Retain grad should be mentioned before backward operation
"""
z.backward()
print(x.grad) ## dz/dx  
print(y.grad) ## dz/dy 
print(z.grad) ## dz/dz

```

```python
from keras.models import Model
from keras.layers import *
```

```python
x=Input(shape=(3,))
```

```python
hidden1=Dense(128,activation="relu")(x) ## hidden1 gets input from x
hidden2=Dense(128,activation="relu")(hidden1)### hidden 2 gets input from hidden1

reg=Dense(1,activation="linear")(hidden2)
clf=Dense(2,activation="sigmoid")(hidden2)
model=Model(inputs=x, outputs=[reg,clf])
```

```python
model.summary()
```

```python
from keras.utils import plot_model

plot_model(model, show_shapes=True, show_layer_names=True)
```

Multiple Inputs and 1 output NN

```python
inputA=Input(shape=(3,))
inputB=Input(shape=(3,))
```

```python
x=Dense(128,activation="relu")(inputA)

y=Dense(128,activation="relu")(inputB)
```

```python
combined=concatenate([x,y])
z=Dense(1,activation="linear")(combined)
```

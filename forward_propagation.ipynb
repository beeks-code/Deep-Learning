```python
import numpy as np
```

<h3> Coding matrix multiplication </h3>

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def get_ouput(weight_matrix,inputs,bias):
    z=np.dot(weight_matrix.T,inputs)+bias
    return sigmoid(z)
```

```python
weight_matrix_1=np.array([[1,2,3],[3,6,-2],[4,-2,4],[-3,2,4]])
input_given=np.array([3,4,5,6])
bias_1=np.array([2,3,4])
```

```python
o_1=get_ouput(weight_matrix_1,input_given,bias_1)
o_1
```

FOR LAYER 2

```python
weight_matrix_2=np.array([[2,3],[3,-2],[-2,9]])
bias_2=np.array([3,4])
o_2=get_ouput(weight_matrix_2,o_1,bias_2)
o_2
```

For Layer 3

```python
weight_matrix_3=np.array([[2],[-2]])
bias_3=np.array([3])
o_3=get_ouput(weight_matrix_3,o_2,bias_3)
o_3
```
